### 1.1 Vectors

  - A vector is an ordered list of numbers. It can represent a point in space, a direction, or various other quantities in machine learning and neural networks. Vectors are fundamental in representing data and performing operations in neural networks.

```python
# creating a vector

# A vector can be created as a 1-dimensional NumPy array.

import numpy as np

# Creating a vector
vector = np.array([1, 2, 3])

print("Vector:", vector)
```

```python
# Vector Addition

# Creating another vector
vector_b = np.array([4, 5, 6])

# Adding two vectors
vector_sum = vector + vector_b
print("Vector Addition:", vector_sum)
```

```python
# Subtracting one vector from another involves subtracting corresponding elements.


# Subtracting two vectors
vector_diff = vector_b - vector
print("Vector Subtraction:", vector_diff)
```

```python
# Scalar Multiplication

# Multiplying a vector by a scalar involves multiplying each element of the vector by that scalar.

scalar = 3
scaled_vector = scalar * vector
print("Scalar Multiplication:", scaled_vector)

```

```python
# Dot Product

# The dot product (or inner product) of two vectors is the sum of the products of their corresponding elements. The dot product is used in many machine learning algorithms, including those for neural networks.


# Dot product of two vectors
dot_product = np.dot(vector, vector_b)
print("vector : ", vector)
print("vector_b: ",vector_b)
print("Dot Product:", dot_product)#4+10+18
```

```python
# Vector Magnitude (Norm)
# The magnitude (or norm) of a vector is a measure of its length. It's computed as the square root of the sum of the squares of its elements.


# Magnitude (or norm) of a vector
magnitude = np.linalg.norm(vector)
print("Vector Magnitude:", magnitude)
```

```python
# Normalization
# Normalization of a vector involves dividing each element by the vector's magnitude to produce a unit vector (a vector with a magnitude of 1).

# Normalizing a vector
normalized_vector = vector / magnitude
print("Normalized Vector:", normalized_vector)


```

### 1.2 Matrices

- Definition: A matrix is a two-dimensional array of numbers, organized in rows and columns. Matrices are used to represent linear transformations, systems of linear equations, and more in neural networks.

- Operations: Learn how to perform matrix addition, subtraction, and multiplication. Matrix multiplication is especially important, as it is a key operation in the feedforward process of neural networks.

```python
# Matrix Creation

import numpy as np

# Create a 2x2 matrix
matrix_a = np.array([[1, 2], [3, 4]])
print("Matrix A:\n", matrix_a)
```

```python
#Matrix Addition

#Matrices of the same size can be added together by adding their corresponding elements.

# Create another 2x2 matrix
matrix_b = np.array([[5, 6], [7, 8]])

# Matrix addition
matrix_sum = matrix_a + matrix_b
print("Matrix A + Matrix B:\n", matrix_sum)
```

```python
#Matrix Subtraction

#Similar to addition, matrix subtraction involves subtracting corresponding elements of two matrices.

# Matrix subtraction
matrix_diff = matrix_a - matrix_b

print("Matrix A - Matrix B:\n", matrix_diff)
```

```python
# Scalar Multiplication

#You can multiply a matrix by a scalar (a single number), which multiplies each element of the matrix by that scalar.

# Scalar multiplication
scalar = 2
matrix_scaled = scalar * matrix_a
print("2 * Matrix A:\n", matrix_scaled)
```

```python
# Element-wise Multiplication

#Also known as the Hadamard product, this operation multiplies corresponding elements of two matrices of the same size.


matrix_hadamard = matrix_a * matrix_b
print("Element-wise multiplication of A and B:\n", matrix_hadamard)
```

```python
# Matrix Multiplication (Dot Product)

#This is a key operation in neural networks, where the rows of the first matrix are multiplied by the columns of the second matrix. The number of columns in the first matrix must match the number of rows in the second matrix.

# Matrix multiplication (Dot Product)
matrix_product = np.dot(matrix_a, matrix_b)
print("Matrix A * Matrix B (Dot Product):\n", matrix_product)
```

```python
# Matrix Transposition

#Transposing a matrix involves flipping it over its diagonal, which switches the row and column indices of the elements.

matrix_transpose = matrix_a.T
print("Transpose of Matrix A:\n", matrix_transpose)
```

```python
# Determinant of a Matrix

#The determinant is a scalar value that is a function of the entries of a square matrix. It can be used to determine whether a matrix has an inverse and other properties.

# Determinant of a matrix
determinant_a = np.linalg.det(matrix_a)
print("Determinant of Matrix A:", determinant_a)
```

```python
# Inverse of a Matrix

#The inverse of a matrix is another matrix such that when multiplied together, they produce the identity matrix. Only square matrices with non-zero determinants have inverses.

matrix_inverse = np.linalg.inv(matrix_a)
print("Inverse of Matrix A:\n", matrix_inverse)
```

```python
# Identity Matrix

# An identity matrix is a square matrix with ones on the diagonal and zeros elsewhere. It acts as the multiplicative identity in matrix multiplication.

# Identity matrix of size 2x2
identity_matrix = np.eye(2)
print("Identity Matrix:\n", identity_matrix)
```

```python
# Matrix Trace

#The trace of a matrix is the sum of the elements on its main diagonal (top-left to bottom-right).


matrix_trace = np.trace(matrix_a)
print("Trace of Matrix A:", matrix_trace)
```

```python
# Example with all the combinations


import numpy as np

# Create matrices
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Scalar multiplication
matrix_scaled = 2 * matrix_a

# Add the scaled matrix to another matrix
result_matrix = matrix_scaled + matrix_b

# Calculate the determinant of the result
determinant_result = np.linalg.det(result_matrix)

# Inverse of the result matrix
inverse_result = np.linalg.inv(result_matrix)

# Trace of the result matrix
trace_result = np.trace(result_matrix)

# Display the results
print("Scaled Matrix A:\n", matrix_scaled)
print("Resulting Matrix after Addition:\n", result_matrix)
print("Determinant of the Result Matrix:", determinant_result)
print("Inverse of the Result Matrix:\n", inverse_result)
print("Trace of the Result Matrix:", trace_result)
```

**Example 2: Eigenvalues and Eigenvectors in a Real-World Application (PCA)**

Eigenvalues and eigenvectors are used in Principal Component Analysis (PCA) to find the principal components of a dataset.


```python
import matplotlib.pyplot as plt

# Generate a 2D dataset
np.random.seed(0)
X = np.random.multivariate_normal([0, 0], [[3, 1], [1, 2]], 100)

# Center the data
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# Covariance Matrix: The covariance matrix captures the variance and covariance of the data. Eigenvectors of this matrix represent the principal directions of the data (principal components), and eigenvalues tell us how much variance exists along these directions.

# Covariance matrix
cov_matrix = np.cov(X_centered.T)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Plotting the data and the principal components
plt.scatter(X_centered[:, 0], X_centered[:, 1])
for eigenvector in eigenvectors.T:
    plt.quiver(0, 0, eigenvector[0], eigenvector[1], angles='xy', scale_units='xy', scale=3, color='r')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title('Principal Components')
plt.grid(True)
plt.show()

# Visualization: The red arrows in the plot represent the principal components. The length of these arrows corresponds to the amount of variance along these directions (scaled by the eigenvalues).
```

### Example: Eigenvalues and Eigenvectors in Neural Networks

Eigenvalues and eigenvectors are not directly used in the typical forward and backward propagation steps of a neural network, but they play a significant role in understanding the properties of the network, especially in terms of stability and optimization. One area where they can be applied is in **weight initialization** and **analyzing the Hessian matrix** during optimization.

#### **Use Case: Weight Initialization with Spectral Norm**

In this example, we will demonstrate how to use the **spectral norm** of a weight matrix in a simple neural network layer. The spectral norm is the largest singular value of a matrix, which is closely related to the largest eigenvalue when the matrix is square. Proper weight initialization is crucial for training deep neural networks effectively.

#### **Step-by-Step Implementation**

1. **Setting up a simple neural network layer:**
   We will define a simple fully connected layer with a weight matrix. Our goal is to initialize the weights using the spectral norm.

2. **Computing the Spectral Norm:**
   We compute the spectral norm (the largest eigenvalue) of the weight matrix to normalize it, ensuring that the network is stable during training.

3. **Using the Spectral Norm for Weight Initialization:**
   We will use the spectral norm to rescale the weights, which can help in maintaining the stability of the gradient during backpropagation.

Let's implement this in Google Colab using Python and NumPy.

#### **1. Defining a Simple Neural Network Layer**

We'll define a single-layer network with a weight matrix and bias.



```python
import numpy as np

# Simple fully connected layer
class SimpleLayer:
    def __init__(self, input_dim, output_dim):
        # Initialize weights randomly
        self.W = np.random.randn(input_dim, output_dim)
        self.b = np.zeros((1, output_dim))

    def forward(self, X):
        # Forward pass
        return np.dot(X, self.W) + self.b

# Example usage
input_dim = 3
output_dim = 2
layer = SimpleLayer(input_dim, output_dim)
```


#### **2. Computing the Spectral Norm**

Next, we'll compute the spectral norm of the weight matrix. The spectral norm is the largest singular value (or the square root of the largest eigenvalue of \( W^T W \) if the matrix is square).


```python

# Compute the spectral norm (largest singular value)
def spectral_norm(W):
    eigenvalues, _ = np.linalg.eig(np.dot(W.T, W))
    return np.sqrt(np.max(eigenvalues))

# Compute the spectral norm of the weight matrix
snorm = spectral_norm(layer.W)
print("Spectral Norm of Weight Matrix:", snorm)
```

#### **3. Normalizing the Weight Matrix**

To ensure that the weight matrix has a controlled norm, which can prevent exploding or vanishing gradients during training, we'll normalize it using the spectral norm.



```python

# Normalize the weight matrix using the spectral norm
layer.W = layer.W / snorm

# Verify the new spectral norm
snorm_normalized = spectral_norm(layer.W)
print("Normalized Spectral Norm of Weight Matrix:", snorm_normalized)

```

#### **4. Forward Pass with Normalized Weights**

Now, let's perform a forward pass using the normalized weight matrix.


```python
# Input data (batch of size 1 for simplicity)
X = np.array([[0.5, 1.0, -1.5]])

# Forward pass
output = layer.forward(X)
print("Output after forward pass:", output)
```


#### **Why This Matters**

- **Stability in Training:** By normalizing the weight matrix with the spectral norm, we ensure that the network's activations do not grow or shrink excessively as they propagate through layers. This can be crucial in deep networks where small changes in weights can lead to significant changes in output.
  
- **Gradient Flow:** Proper initialization and normalization help maintain a stable gradient flow during backpropagation, which is vital for effective training, especially in deep networks.

#### **Conclusion**

In this example, we used the concept of eigenvalues to compute the spectral norm of a weight matrix in a simple neural network layer. By normalizing the weights with this spectral norm, we aimed to improve the stability of the network during training. While this is a relatively simple example, the underlying concept is powerful and widely used in more advanced techniques, such as spectral normalization in Generative Adversarial Networks (GANs) and in controlling the Lipschitz constant of the network.

Eigenvalues and eigenvectors provide valuable insights into the properties of matrices used in neural networks, allowing us to design more robust and stable models.

#### Application in Neural Networks: Backpropagation

1. **Imports:**
   - `numpy` is imported as `np`. This library is used for handling arrays and performing matrix operations efficiently.



```python
import numpy as np
```

2. **Activation Function (Sigmoid):**
   - The `sigmoid` function is used to introduce non-linearity into the network. It's crucial for enabling the network to learn complex patterns.
   - The derivative of the sigmoid function (`sigmoid_derivative`) is necessary for backpropagation, allowing the network to update weights effectively.


```python
# Sigmoid function: This is the activation function used in our neural network.
# It maps the input to a value between 0 and 1.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the Sigmoid function: This is used during backpropagation to compute the gradient.
# It helps in determining how much the output of the sigmoid function changes with respect to its input.
def sigmoid_derivative(x):
    return x * (1 - x)

```


3. **Dataset:**
   - The input dataset is a simple XOR problem, where each input is a pair of binary values (0 or 1).
   - The expected output is the result of the XOR operation applied to the inputs.


```python

# Input data: This is a simple XOR dataset with 4 samples, each having 2 features.
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Expected output: The correct outputs for the XOR problem, where:
# - 0 XOR 0 = 0
# - 0 XOR 1 = 1
# - 1 XOR 0 = 1
# - 1 XOR 1 = 0
expected_output = np.array([[0], [1], [1], [0]])
```


4. **Network Architecture:**
   - The network has an input layer with 2 neurons (for the two features of the input).
   - It has a hidden layer with 2 neurons (a design choice).
   - The output layer has 1 neuron, providing a single output for the XOR result.


```python

# Initialize network parameters:
# - input_layer_neurons: Number of neurons in the input layer (2 features).
# - hidden_layer_neurons: Number of neurons in the hidden layer (chosen as 2 for simplicity).
# - output_neurons: Number of neurons in the output layer (1 output).

input_layer_neurons = inputs.shape[1]  # This is 2, because each input has 2 features.
hidden_layer_neurons = 2  # We chose to have 2 neurons in the hidden layer.
output_neurons = 1  # We have 1 output for each input.

# Randomly initialize weights for the connections:
# - weights_input_hidden: Weights between input layer and hidden layer.
# - weights_hidden_output: Weights between hidden layer and output layer.
```


5. **Weights Initialization:**
   - Weights are initialized randomly. The weights will be adjusted during training to minimize the error between the predicted output and the expected output.


```python

# np.random.uniform generates random numbers between 0 and 1.
weights_input_hidden = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
weights_hidden_output = np.random.uniform(size=(hidden_layer_neurons, output_neurons))
```


6. **Learning Rate:**
   - The learning rate determines the step size during the weight updates. A smaller learning rate might lead to slower but more stable learning.


```python

# Learning rate: This controls how much the weights are adjusted with respect to the gradient.
learning_rate = 0.1
```


7. **Training Loop (10,000 Epochs):**
   - **Forward Propagation:**
     - The inputs are fed into the network, and the outputs of the hidden and output layers are calculated using matrix multiplication and the sigmoid activation function.
   - **Error Calculation:**
     - The difference between the predicted output and the actual output is computed, which will be minimized by adjusting the weights.
   - **Backpropagation:**
     - Gradients are calculated using the chain rule, which involves computing how the error changes with respect to each weight.
     - The weights are updated using these gradients, effectively "learning" from the error.
   - **Weight Updates:**
     - The weights connecting the hidden to the output layer and the input to the hidden layer are updated based on the gradients and the learning rate.


```python

# Training loop: We will train the network for 10,000 epochs (iterations).
for epoch in range(10000):
    # Forward Propagation: Calculating the output of the network.

    # Step 1: Compute the input to the hidden layer (activation of hidden layer).
    # This is done by multiplying the inputs with the weights connecting input to hidden layer.
    hidden_layer_activation = np.dot(inputs, weights_input_hidden)

    # Step 2: Apply the sigmoid function to the hidden layer activation.
    # This gives us the output of the hidden layer.
    hidden_layer_output = sigmoid(hidden_layer_activation)

    # Step 3: Compute the input to the output layer (activation of output layer).
    # This is done by multiplying the hidden layer output with the weights connecting hidden to output layer.
    output_layer_activation = np.dot(hidden_layer_output, weights_hidden_output)

    # Step 4: Apply the sigmoid function to the output layer activation.
    # This gives us the predicted output of the network.
    predicted_output = sigmoid(output_layer_activation)

    # Calculate error: This is the difference between the expected output and the predicted output.
    # This error will be used to adjust the weights during backpropagation.
    error = expected_output - predicted_output

    # Backpropagation: Adjusting the weights to minimize the error.

    # Step 5: Calculate the gradient for the output layer.
    # This involves the error and the derivative of the sigmoid function at the output.
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    # Step 6: Calculate the error in the hidden layer.
    # This is done by multiplying the gradient at the output with the transposed weights of the hidden-to-output layer.
    error_hidden_layer = d_predicted_output.dot(weights_hidden_output.T)

    # Step 7: Calculate the gradient for the hidden layer.
    # This involves the hidden layer error and the derivative of the sigmoid function at the hidden layer.
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)

    # Step 8: Update the weights for the connections between the hidden and output layers.
    # Weights are updated by adding the product of the learning rate and the gradient.
    weights_hidden_output += hidden_layer_output.T.dot(d_predicted_output) * learning_rate

    # Step 9: Update the weights for the connections between the input and hidden layers.
    # Similarly, update these weights using the gradients calculated for the hidden layer.
    weights_input_hidden += inputs.T.dot(d_hidden_layer) * learning_rate
```


8. **Final Output:**
   - After training, the network's final output for the XOR problem is printed. Ideally, it should closely approximate the expected XOR results.


```python

# After the training loop, the network has learned to approximate the XOR function.
# Let's print the final output from the network after training.
print("Final output after training:\n", predicted_output)
```
